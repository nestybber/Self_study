```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset_df = pd.read_csv('./customer_satisfaction/train.csv')
print('Dataset shape: ', dataset_df.shape)
dataset_df.head()
```

```python
dataset_df.info()
```

```python
dataset_df['TARGET'].value_counts()
```

```python
# 0은 만족, 1은 불만족. 불만족의 비율이 현저히 적으므로 evaluation시 accuracy가 아닌 ROC_AUC이용
dataset_df.describe()
```

```python
print(dataset_df['var3'].value_counts()[:10])
```

```python
# -999999값이 116개 들어있어 적절한 값으로 encoding해주기 (가장 많이있는 값인 2로 대체), ID는 필요없어 제거
dataset_df['var3'].replace(-999999, 2, inplace=True)
dataset_df.drop('ID', axis=1, inplace=True)
```

```python
# feature set과 target set 분리
X_features = dataset_df.iloc[:, :-1]
y_labels = dataset_df.iloc[:, -1]
```

```python
# train set과 test set 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size=0.2)
train_cnt = y_train.count()
test_cnt = y_test.count()
print('Train set label value distribution ##')
print(y_train.value_counts()/train_cnt)
print('\nTest set label value distribution ##')
print(y_test.value_counts()/test_cnt)
```

```python
# 0과 1이 비슷한 비율로 분리되었음을 확인
# 1. XGBoost로 학습 모델 생성
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb_clf = XGBClassifier(n_estimators=500)
xgb_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="auc", eval_set=[(X_train, y_train), (X_test, y_test)])
# roc_auc score는 input으로 test set label값과 model의 test set의 1 예측확률 값을 필요로 함
xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1], average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

# hyper parameter test 수행속도 향상 위해 estimator 개수 줄임
xgb_clf = XGBClassifier(n_estimators=100)

params = {'max_depth':[5,7], 'min_child_weight':[1,3], 'colsample_bytree':[0.5,0.75]}

grid_cv = GridSearchCV(xgb_clf, param_grid=params)
grid_cv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric='auc', eval_set=[(X_train, y_train), (X_test, y_test)])

print('GridSearchCV best parameter: ', grid_cv.best_params_)
xgb_roc_score = roc_auc_score(y_test, grid_cv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
from lightgbm import LGBMClassifier
lgbm_clf = LGBMClassifier(n_estimators=500)

lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_test, y_test)], verbose=True)
lgbm_roc_score = roc_auc_score(y_test, lgmb_clf.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```
