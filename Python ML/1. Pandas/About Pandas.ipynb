```python
import pandas as pd
import numpy as np

# reading csv file and make dataframes
titanic_df = pd.read_csv('train.csv')
titanic_df
```

```python
# list or ndarray to dataframe
col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
df_list1 = pd.DataFrame(list1, columns=col_name1)
df_array1 = pd.DataFrame(array1, columns=col_name1)
print(df_list1)
```

```python
# dictionary to dataframe
# key -> column name, value -> data
dict = {'col1':[1, 11], 'col2':[2, 22]}
df_dict = pd.DataFrame(dict)
print(df_dict)
```

```python
# dataframe to ndarray, list and dict
arr = df_dict.values
print(arr, type(arr))
```

```python
# adding series into dataframe
titanic_df['New'] = 0
titanic_df.head(3)
```

```python
# deleting series of dataframe
titanic_df_dr = titanic_df.drop('New', axis = 1)
titanic_df_dr.head(3)
```

```python
# extracting indexes of df
indexes = titanic_df.index
print(indexes)
print(indexes.values)
```

```python
# filtering
titanic_boolean = titanic_df[titanic_df['Age']>70]
titanic_boolean
```

```python
titanic_df[titanic_df['Age']>60][['Name', 'Age']].head(3)
```

```python
titanic_df.loc[titanic_df['Age']>60, ['Name', 'Age']].head(3)
```

```python
# sorting
titanic_sorted = titanic_df.sort_values(by=['Pclass', 'Name'], ascending = False)
titanic_sorted.head(3)
```

```python
# take a mean value
titanic_groupby = titanic_df.groupby('Pclass')
titanic_groupby[['PassengerId' , 'Survived']].mean()
```

```python
# applying lambda function
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x < 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(3)
```
