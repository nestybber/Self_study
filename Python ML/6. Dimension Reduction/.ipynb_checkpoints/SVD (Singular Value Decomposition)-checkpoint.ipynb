```python
# Numpy에 svd모듈 존재
import numpy as np
from numpy.linalg import svd

np.random.seed(121)
a = np.random.randn(4, 4)
print(np.round(a,3))
```

```python
U, Sigma, Vt = svd(a)
```

```python
print(U.shape, Sigma.shape, Vt.shape)
print('U matrix: \n', np.round(U, 3))
print('Sigma value: \n', np.round(Sigma, 3))
print('V transpose matrix: \n', np.round(Vt, 3))
```

```python
Sigma_matrix = np.diag(Sigma)
a_ = np.dot(np.dot(U, Sigma_matrix), Vt)
print(np.round(a_,3))
```

```python
# 위의 경우는 a의 모든 row vextor가 일차 독립이기 때문에 sigma 값이 모두 존재하였음.
# 일차 종속관계가 있을시, 그 row vector에 대해서는 sigma 값 존재 X

a[2] = a[0] + a[1]
a[3] = a[0]
print(np.round(a, 3))

U, Sigma, Vt = svd(a)
print('Sigma value: \n', np.round(Sigma, 3))
```

```python
# 복원 시 앞 2열,행씩만 추출
U_ = U[:,:2]
Vt_ = Vt[:2, :]
Sigma_ = np.diag(Sigma[:2])

a_ = np.dot(np.dot(U_, Sigma_), Vt_)
print(np.round(a_, 3))
```

```python
# Scikit learn의 TruncatedSVD Class 이용
from sklearn.decomposition import TruncatedSVD, PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris_data = iris.data

tsvd = TruncatedSVD(n_components=2)
tsvd.fit(iris_data)
iris_tsvd = tsvd.transform(iris_data)

plt.scatter(x=iris_tsvd[:,0], y=iris_tsvd[:,1], c=iris.target)
plt.xlabel('trunSVD comp1')
plt.ylabel('trunSVD comp2')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris_data)

# Truncated SVD
tsvd = TruncatedSVD(n_components=2)
tsvd.fit(iris_scaled)
iris_tsvd = tsvd.transform(iris_scaled)

# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(iris_scaled)
iris_pca = pca.transform(iris_scaled)

# 각 변환 결과 출력하기
fig, (ax1, ax2) = plt.subplots(figsize=(9, 4), ncols=2)
ax1.scatter(x=iris_tsvd[:,0], y=iris_tsvd[:,1], c=iris.target)
ax2.scatter(x=iris_pca[:,0], y=iris_pca[:,1], c=iris.target)
ax1.set_title('Truncated SVD Transformed')
ax2.set_title('PCA Transformed')
```

```python
# 결과값 거의 동일!
# Scaled된 dataset에 대해서는 scikit learn의 SVD와 PCA는 동일 변환 수행
# 단, PCA는 Dense Matrix만 변환 가능하지만 SVD는 Sparse Matrix도 변환 가능
```
