```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df.head()
```

```python
iris_df['target'].value_counts()
```

```python
iris_df.info()
```

```python
markers = ['^', 's', 'o']
for i, marker in enumerate(markers):
    x_axis_data = iris_df[iris_df['target'] == i]['sepal length (cm)']
    y_axis_data = iris_df[iris_df['target'] == i]['sepal width (cm)']
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label=iris.target_names[i])
    
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
# PCA는 여러 속성의 값을 연산해야 하므로 속성의 scale에 영향을 받음 -> Scaling 사전에 한 후 PCA에 적용
from sklearn.preprocessing import StandardScaler

iris_scaled = StandardScaler().fit_transform(iris_df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(iris_scaled)
iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
pca_col = ['pca_component_1', 'pca_component_2']
iris_df_pca = pd.DataFrame(iris_pca, columns=pca_col)
iris_df_pca['target'] = iris.target
iris_df_pca.head()
```

```python
markers = ['^', 's', 'o']
for i , marker in enumerate(markers):
    x_axis_data = iris_df_pca[iris_df_pca['target'] == i]['pca_component_1']
    y_axis_data = iris_df_pca[iris_df_pca['target'] == i]['pca_component_2']
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label=iris.target_names[i])
    
plt.legend()
plt.xlabel('comp1')
plt.ylabel('comp2')
plt.show()
    
```

```python
# 전체 변동성에서 개별 PCA component별 차지하는 변동성 비율 확인
print(pca.explained_variance_ratio_)
```

```python
# 1번째 pca요소인 pca_component_1이 전체 변동성의 약 76.7% 차지, 2번째 요소가 18.3% 차지
# 즉 2개의 요소로도 원본 데이터 변동성의 95% 정도 설명 가능
```

```python
df = pd.read_excel('credit_card.xls', sheet_name='Data')
df.head()
```

```python
df.shape
```

```python
df.rename(columns={'PAY_0':'PAY_1', 'default payment next month':'default'}, inplace=True)
y_target = df['default']
X_data = df.drop('default', axis=1, inplace=False)
```

```python
import seaborn as sns
corr = X_data.corr()
# 각 속성간의 상관도 구하기
plt.figure(figsize=(14,14))
sns.heatmap(corr, annot=True, fmt='.1g')
```

```python
# BILL_AMT 속성 간 상관도 매우 높고, PAY 속성도 서로간의 상관도 높음 -> PCA변환
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cols_bill = ['BILL_AMT'+str(i) for i in range(1, 7)]
print('feature names: ', cols_bill)

scaler = StandardScaler()
df_cols_scaled = scaler.fit_transform(X_data[cols_bill])
pca = PCA(n_components = 2)
pca.fit(df_cols_scaled)
print('Variance Ratio : ', pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rcf = RandomForestClassifier(n_estimators=300)

scaler = StandardScaler()
df_scaled = scaler.fit_transform(X_data)

pca = PCA(n_components=7)
df_pca = pca.fit_transform(df_scaled)
scores_pca = cross_val_score(rcf, df_pca, y_target, scoring='accuracy', cv=3)

print('Fold set accuracy: {0:.4f}'.format(scores_pca))
```
